```python
from IPython.display import Image as IPythonImage
from PIL import Image, ImageFont, ImageDraw
```

```python
def display_cover(top, bottom):
    
    import requests
    
    name = "album_art_raw.png"
    album_art_raw = requests.get("https://picsum.photos/777/777/?random")
    # "wb" for binary read, write access - truncates file to 0 bytes
    with open(name, "wb") as album_art_raw_file:                         #returns file object
        album_art_raw_file.write(album_art_raw.content)                  #content response in bytes
    
    img = Image.open("album_art_raw.png")
    draw = ImageDraw.Draw(img)
    
    band_name_font = ImageFont.truetype("/Library/Fonts/Luminari.ttf", 42)    #load Font
    album_name_font = ImageFont.truetype("/Library/Fonts/Luminari.ttf", 27)
    
    band_x, band_y = 50, 50
    album_x, album_y = 50, 700
    
    outline_color = "black"
    draw.text((band_x-1, band_y-1), top, font=band_name_font, fill=outline_color)
    draw.text((band_x+1, band_y-1), top, font=band_name_font, fill=outline_color)
    draw.text((band_x-1, band_y+1), top, font=band_name_font, fill=outline_color)
    draw.text((band_x+1, band_y+1), top, font=band_name_font, fill=outline_color)

    draw.text((album_x-1, album_y-1), bottom , font=album_name_font, fill=outline_color)
    draw.text((album_x+1, album_y-1), bottom , font=album_name_font, fill=outline_color)
    draw.text((album_x-1, album_y+1), bottom , font=album_name_font, fill=outline_color)
    draw.text((album_x+1, album_y+1), bottom , font=album_name_font, fill=outline_color)
    
    draw.text((band_x,band_y),top,(255,255,255),font=band_name_font)
    draw.text((album_x, album_y),bottom,(255,255,255),font=album_name_font)
    
    return img
    
```

```python
img = display_cover(top="Never", bottom="give up")
img.save("sample_out.png")
img2 = display_cover(top="Python", bottom="Data Science")
img2.save("sample_out2.png")
```
