```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://de.wikipedia.org/wiki/Liste_der_Stra%C3%9Fennamen_in_Ravensburg"
requests_object = requests.get(url).text
soup = BeautifulSoup(requests_object,"lxml")
#print(soup.prettify())
```

```python
wikitable = soup.find("table",{"class":"wikitable sortable zebra"})
wikitable
```

```python
streets = []
tds = wikitable.findAll("td").text
#links
for td in tds:
    streets.append(td.get("td"))
print(streets)
```
